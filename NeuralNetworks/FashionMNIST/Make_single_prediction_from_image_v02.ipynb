### Convert image into an array

```python
from tensorflow.keras.preprocessing.image import img_to_array

# convert to array
img = img_to_array(img)
```

### Transforming image to one single channel one

```python
# reshape into a single sample with 1 channel
img = img.reshape(1, 28, 28, 1)
```

### Scaling data
The model is trained with min-max scaling

```python
# prepare pixel data
img = img.astype('float32')
img = img / 255.0
```

```python
from tensorflow.keras.preprocessing.image import load_img

model_name = '/content/drive/MyDrive/PIA/Tareas/Tareas_Redes_Neuronales/Zalando \
/Files/final_model_fashion_mnist_final_83.h5'
model = load_model(model_name)
```


> * [Multiclass label prediction in Keras sequential model](https://datascience.stackexchange.com/questions/79761/class-label-prediction-in-keras-sequential-model-showing-different-results-in-co)

```python
import numpy as np

# predict the class
result = np.argmax(model.predict(img), axis=-1)
print(f'Image class is: {result}')
```
