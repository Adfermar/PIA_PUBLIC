### STRINGS

1. Son cadenas de texto encerradas entre comillas dobles o simples (línea simple) o bien entre comillas triples para múltiples líneas:
> * `"cadena de texto"`
> * `'cadena de texto'`
> * `"""cadena de texto`
> * `en múltiples líneas"""`
1. Se codifican en **Unicode**
1. Una vez que se asigna una cadena de texto a una variable ésta **NO puede cambiarse**
1. Se accede a una posición de una variable tipo STRING de la siguiente forma:
> * `str = "hola mundo"`
> * `print(str[0])`
> * La primera posición de una cadena es cero y la última es (n-1) siendo n la longitud de la cadena.
> * Se puede acceder a la posición de una cadena de forma inversa: "-1" es la última posición mientras que (-n) sería la primera.
> * Podemos conocer la longitud de una cadena con la función `len()`
1. Se puede acceder a partes de una cadena con el formato: `cadena[inicio:fin:paso]`
1. Casos particulares útiles:
> * `print(my_string[:])`  muestra la cadena entera
> * `print(my_string[::-1])`  invierte la cadena al completo
1. Introducir números u otro texto dentro de una cadena de texto:
> * Con el operador `%s` introducimos una cadena de texto.
> * Con el operador `%i`introducimos un número entero.
> * Con el operador `%f` introducimos un núermo decimal. Podemos limitar en número de decimales con `%f.2f`, en este caso, dos decimales.


````python
# Definir una variable de tipo cadena:
str1 = "hola"
print(str1)

# Acceder a la primera posición de la cadena:
print(str1[0])

# Longitud de la cadena:
n = len(str1)
print("la longitud de la cadena es:", n)

# Acceder a la última posición de la cadena:
print(str1[n-1])

# Acceder a la última posición utilizando la indexación inversa:
print(str1[-1])

# Acceder a la última posición utilizando la indexación inversa:
print(str1[-n])

# Muestra la cadena al completo
print(str1[:])

# Muestra la cadena en orden inverso
print(str1[::-1])

# Meter una cadena de texto dentro de otra
str2 = "Hola Mundo %s" % ("Cruel")
print(str2)

# Meter un número entero dentro de una cadena de texto
str2 = "Mi sueldo es: %i euros al mes" % (1000)
print(str2)

# Meter varios números enteros dentro de una cadena de texto
str2 = "Números impares: %i, %i, %i" % (1, 3, 5)
print(str2)

# Limitar el número de decimales al meter un número real dentro de una cadena
str2 = "%.2f" % (1.112233)
print(str2) # '1.11'
````

## OPERADORES


---


Los operadores pueden colocarse ANTES del operando o ENTRE ellos.

### ARITMÉTICOS
> Se ejecutan por orden de precedencia. En caso de que ésta sea la misma el orden de ejecución se determinará **de izquierda hacia derecha**
> 1. Paréntesis `()`
> 1. Exponente `**`
> 1. Resto `%`; multiplicación `*`; división de números reales `/`; si queremos que una división que resulta en un núermo real devuelva un número **ENTERO** utilizamos el operador: `//`
> 1. Suma `+`; resta `-`

````python
# Ejemplo de precedencia entre operadores
# Primero se realiza la multiplicación y lueg la resta (distinto nivel de precedencia)
print(10 - 3 * 2) 

# Operadores con la misma precedencia
# Primero se ejecuta la multiplicación (por estar a la izquierda) y luego la división
print(3 * 20 / 5)  
 # División primero, luego multiplicación
print(3 / 20 * 5) 

# Utilización de paréntesis
# Todas las operaciones entre paréntesis se ejecutan primero empezando desde
# la izquierda hacia la derecha

# Primero la resta y luego la multiplicación
print((10 - 3) * 2) 
# Primero la suma, luego el resto y por último la división
print((18 + 2) / (10 % 8))
````

### COMPARADORES

Devuelven un tipo **boolenano** que puede ser _True_ o _False_

> 1. `>` `<` `>=` `<=` `==` `!=`
> 1. Los operadores `is` y `is not` NO son lo mismo que `==` y `!=`. Los primeros permiten distinguir entre _elementos_ distintos (por ejemplo variables) **independientemente de su contenido**

````python
# Ejemplos con operadores == y != y su diferencia con "is" e "is not"

lista1 = [1, 2, 3]
lista2 = [1, 2, 3]
lista3 = [1, 2, 4]
# Lista 1 y 2 tiene el mismo contenido
print(lista1 == lista2)
# lista 1 y lista 2 NO son el mismo elemento
print(lista1 is lista2)

# Lista 2 y 3 NO tienen el mismo contenido
print(lista2 == lista3)
# Lista 2 y 3 tampoco son el mismo elemento. "is not" devolverá True
print(lista2 is not lista3)

# Veamos lo que ocurre cuando los elementos son números:
num1 = 1
num2 = 1
print (num1 is num2)
# Vemos que ahora las variables tienen el mismo elemento
````

### RECURSIVIDAD

1. Podemos utilizar recursividad en Python. La mejor forma de explicarlo es con un ejemplo:


````python
# Función recursiva que muestra por pantalla un número decreciente cada vez:
def rec_count(number):
    print(number)
    # Base case
    if number == 0:
        return 0
    rec_count(number - 1)  # A recursive call with a different argument

rec_count(5)

print("-------------------------------------------------")

# Otro ejemplo de recursividad:
def factorial(n):
    # Base case
    if n == 0 or n == 1:
        return 1

    if n < 0:
        return -1
    # Recursive call
    return n * factorial(n - 1)

result = str(factorial(5))
print("El factorial de 5 es: " + result)
````

> * Propiedades de los conjuntos:
>> * Propiedad **asociativa**: La intersección y la unión de conjuntos satisfacen la propiedad asociativa.
(A⋂B)⋂C = A⋂(B⋂C)    (A⋃B)⋃C = A⋃(B⋃C)
>> * Propiedad **distributiva** La intersección y la unión de conjuntos satisfacen la propiedad distributiva.
A⋃(B⋂C) = (A⋃B)⋂(A⋃C)      A⋂(B⋃C) = (A⋂B)⋃(A⋂C)
>> * **Identidad** A⋃∅ = A A⋂U = A
>> * **Complemento**: A⋃AC = U A⋂AC = ∅
>> * **Idempotente**: A⋂A = A A⋃A = A
>> * Propiedad **conmutativa**. La intersección y la unión de conjuntos satisfacen la propiedad conmutativa.La intersección y la unión de conjuntos satisfacen la propiedad conmutativa.  A⋂B = B⋂A  A⋃B = B⋃A
> * Operaciones con conjuntos:
>> * Unión: operador `|` o método `.union(conunto)`
```python
set_A = {1, 2, 3, 4}

print(set_A | set_B)
print(set_A.union(set_B))
```
>> * Diferencia: operador `-` o método `.difference(conjunto)`. La diferencia no cumple la propiedad conmutativa. No es lo mismo A-B que B-A.
```python
set_A = {1, 2, 3, 4}
set_B = {2, 8, 4, 16}
print(set_A - set_B)
print(set_A.difference(set_B))
print(set_B - set_A)
print(set_B.difference(set_A))
```
>> * Intersección: operador `&` o método `.intersection(conjunto)`
```python
set_A = {1, 2, 3, 4}
set_B = {2, 8, 4, 16}

print(set_A.intersection(set_B))
print(set_B.intersection(set_A))
```

````python
""" Un caso muy curioso del uso de funciones lambda y de la función filter:
---------------------------------------------------------------------------
You must implement the count_low_high() function. 
Its parameter is a list of numbers.

If a number is greater than 50 or divisible by 3, it will count as a high. 
If these conditions are not met, the number is considered a low.

At the end of the function, you must return a list that contains the number 
of lows and highs, in that order.

In case the list is empty, you may return None.

Sample Input#
num_list = [20, 9, 51, 81, 50, 42, 77]
Sample Output#
[2, 5] # 2 lows and 5 highs
"""

def count_low_high(num_list):
    if (len(num_list)==0):
        return None
    high_list = list(filter(lambda n: n > 50 or n % 3 == 0, num_list))
    low_list = list(filter(lambda n: n <= 50 and not n % 3 == 0, num_list))
    return [len(low_list), len(high_list)]


num_list = [20, 9, 51, 81, 50, 42, 77]

print(count_low_high(num_list))
````

### Image.mode
The mode of an image is a string which defines the type and depth of a pixel in the image. Each pixel uses the full range of the bit depth. So a 1-bit pixel has a range of 0-1, an 8-bit pixel has a range of 0-255 and so on. The current release supports the following standard modes:

        1 (1-bit pixels, black and white, stored with one pixel per byte)

        L (8-bit pixels, black and white)

        P (8-bit pixels, mapped to any other mode using a color palette)

        RGB (3x8-bit pixels, true color)

        RGBA (4x8-bit pixels, true color with transparency mask)

        CMYK (4x8-bit pixels, color separation)

        YCbCr (3x8-bit pixels, color video format)

            Note that this refers to the JPEG, and not the ITU-R BT.2020, standard

        LAB (3x8-bit pixels, the L*a*b color space)

        HSV (3x8-bit pixels, Hue, Saturation, Value color space)

        I (32-bit signed integer pixels)

        F (32-bit floating point pixels)
