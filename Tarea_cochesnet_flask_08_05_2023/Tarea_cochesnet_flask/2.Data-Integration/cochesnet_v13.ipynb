````python
def featureDistribution(df, feature):

    from collections import Counter

    values = df[feature].values
    counter = Counter(values)
    for k, v in counter.items():
        p = v / len(values) * 100
        print(f'Class: {k}, Count: {v}, Percentage: {round(p, 2)}%')
````

````python
from unidecode import unidecode
import re

def cleanString(input:str):
    output = input
    # 
    pattern = r'(?:\\u00)'
    reg = re.compile(pattern)    
    if reg.search(input): output =  output.encode().decode('unicode_escape')
    else: output =  output.encode().decode()    
    # Remove leading and talining white spaces (trimming)
    # Replacing inner white spaces by '_'
    # Set all characters to lower-case
    output = output.strip().replace(' ', '_').replace('-', '_').lower()
    # Replacing accented characters by non-accented ones
    output = unidecode(output)
    output = ''.join(['a' if ord(c) >= 128 else c for c in output])
    # Removing all non-ascii character from string
    output =  output.encode('ascii', 'ignore').decode('ascii')
    # Keeping only alpha-numeric characters plus '_' in string
    output = re.sub(r'[^\w\s|_]+', '', output)

    return output
````

````python
def replaceString(input:str):
    input.replace( 
                {'color': r'^commoncolor_', 
                'transmissionType': r'^commontransmission_', 
                'bodyType': r'^body_type_',
                'seller_type': 'profesional',
                'seller_type': 'particular'}, 
                {'color': '', 
                'transmissionType': '',
                'bodyType': '',
                'seller_type': '',
                'seller_type': 'private'}, 
                regex=True,
                inplace=True)
    return input
````

````python
import json

def load_data_from_files(file_list:list, PATH_BASE:str, removeFirstItemsCount=0, property_to_spread=''):
    """
    @file_list <list> -input- >> name list of all JSON files to load.
    @PATH_BASE <string> -input- >> the path of the directory which contains
        all the JSON files.
    @removeFirstItemsCount <numeric> -input- >> number of records in the 
        beginning that will be skipped from each input JSON file.
    @property_to_spread <string> -input- >> field that contains the main
        dictionary.
    @output <list> -output- >> stores in a list all 
        JSON records from ecah file in the list.
    """
    output = list()

    # Iterate over all full paths
    for file_name in file_list:
        # Getting the full file path
        path = PATH_BASE + file_name
        print(path)
        # Open each JSON file
        with open(path, 'r') as f:
            file_data = json.load(f)
            # Skipping the first "removeFirstItemsCount" 
            # records by dropping them
            for i in range(0, removeFirstItemsCount):
                file_data.pop(0)
            # Getting the main dictionary skipping the
            # "property_to_spread" record if exists
            if property_to_spread != '':
                # Add every JSON record to temp. dictionary
                for i, item in enumerate(file_data):
                    resto_de_datos = item.pop(property_to_spread)
                    file_data[i] = {**item, **resto_de_datos}
            print(len(file_data))
            # Add the temp. dictionary to the total dictionary
            output += file_data
    return output
````

````python
# Load or save text files

def loadTextFileFromDisk(PATH):
    with open(PATH, encoding='utf-8', mode = 'r') as f:
        data = f.read()
    f.close()
    return data

def saveToTextFile(PATH, data):
    with open(PATH, encoding='utf-8', mode = 'w') as f:
        f.write(data)
    f.close()

# -----------------------------------------------------

# Load or save files

def loadFromDisk(PATH):
    with open(PATH, mode = 'r') as f:
        data = f.read()
    f.close()
    return data

def saveToDisk(PATH, data):
    with open(PATH, mode = 'w') as f:
        f.write(data)
    f.close()
````

````python
# Load coches net json file removing fields

import json

def open_json_data(filename, removeFirstItemsCount = 0, property_to_spread = ''):
    
    try:
      with open(filename, 'r') as file:
        data = json.load(file)
    except UnicodeDecodeError as UDE:
      raise ValueError(f'{UDE}')
    except Exception as Err:
      print(f'{Err}')

    for i in range(0, removeFirstItemsCount):
      data.pop(0)

    if property_to_spread != '':
      for i, item in enumerate(data):
        resto_de_datos = item.pop(property_to_spread)
        data[i] = {**item, **resto_de_datos}

    return data
````

## Function that perform all cleaning process

````python
def cleaningProcess(df_input:pd.DataFrame):

    # Dummy dataframe
    df_out = df.copy()
    # Dropping specified columns from dataframe
    columns_to_drop = [
        'url', 'datetime', 'fuelTypeId', 'id', 'makeId',
        'modelId', 'provinceId', 'title', 'transmissionTypeId',
        'brand', 'brand_id', 'model_id', 'fuel_types'
    ]
    df_out.drop(columns=columns_to_drop, inplace=True)
    # Process NA/NaN/NAN/null/Na values
    # First option: replace missing values by centrality statistic like 'mode' 
    """
    values = {
        'cubicCapacity':int(round(df_out['cubicCapacity'].mode()[0], 0)),
        'hp':int(round(df_out['hp'].mode()[0], 0)),
        'bodyType':str(df_out['bodyType'].mode()[0]),
    }
    """
    # Second option: replace missing values by 'null' strint to be able
    # to process them later.
    values = 'null'
    df_out.fillna(value=values, inplace=True)
    # Third option:
    # Drop rows that contains missing values
    # df_out.dropna(inplace=True)
    # Drop duplicated rows
    df_out.drop_duplicates(inplace=True)
    # Reset dataframe index
    df_out.reset_index(drop=True, inplace=True)
    # Trying to cast automatically every column to its best data type
    # (It's works from time to time)
    df_out = df_out.infer_objects()
    df_out = df_out.convert_dtypes()
    CATEGORICAL_FEATURES = [
        'color',
        'fuelType',
        'make',
        'model',
        'province',
        'transmissionType',
        'seller_type',
        'bodyType'
    ]
    NUMERIC_FEATURES = ['km', 'cubicCapacity', 'doors', 'hp']
    # Transforming each categorical value by applying "cleanString" function
    df_out.loc[:, CATEGORICAL_FEATURES] = \
        df_out[CATEGORICAL_FEATURES].applymap(lambda x: cleanString(str(x)))
    # Remove "commoncolor_" string in "color" feature
    # Remove "commontransmission_" string in "transmissionType" feature
    # Replace some english expressions found in "seller_type" feature
    # Remove "body_type_" sub-string from "bodyType" feature
    df_out = replaceString(df_out)  
    # Infer data types
    df_out = df_out.infer_objects()
    df_out = df_out.convert_dtypes()
    # Reset dataframe index
    df_out.reset_index(drop=True, inplace=True)

    return df_out, CATEGORICAL_FEATURES, NUMERIC_FEATURES
````
