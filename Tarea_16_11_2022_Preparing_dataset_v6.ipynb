# Enunciado de las tareas

1. Delete the column ‘customerid’ from the dataset.

```python
df = df_original
```



---


## Tarea 1

Delete the column ‘customerid’ from the dataset.

```python
# Tarea 1
df = df.drop("customerID", axis=1)
df.head()
```



---


## Tarea 2

Convert the column ‘TotalCharges’ to a numeric format, coercing errors if necessary.

```python
# Tarea 2
# Convert the column ‘TotalCharges’ to a numeric format, 
# coercing errors if necessary.

df["TotalCharges"].fillna(0, inplace = True)

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
print(type(df["TotalCharges"][0]))
df["TotalCharges"].head(10)
```



---


## Tarea 3

Change all column names to lowercase.

```python
# Tarea 3
# Change all column names to lowercase.
df = df.rename(str.lower, axis='columns')
for i in df.columns:
  print(i)
```



---


## Tarea 4

For string values change spaces to an underscore symbol.

```python
# 4. For string values change spaces to an underscore symbol.
for i in df.index:
  for j in df.columns:
    if( type( df.at[i, j] ) == str ):
      df.at[i, j] = df.at[i, j].replace(' ', '_')
      # También cambiamos el guión medio por el "underscore"
      df.at[i, j] = df.at[i, j].replace('-', '_')
      
```

## Tarea 5

Convert the column ‘churn’ to integer format (no - 0, yes - 1).

```python
# 5. Convert the column ‘churn’ to integer format (no - 0, yes - 1).
for i in df.index:
    if ( df.at[i, "churn"]  == "Yes" ):
      df.at[i, "churn"] = 1
    elif ( df.at[i, "churn"]  == "No" ):
      df.at[i, "churn"] = 0
    else:
      raise ValueError("Wrong value in churn column position:")
```

### Cambiar el tipo de dato de la columna "churn"
